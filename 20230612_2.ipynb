```python
## 텐서플로우 모듈 임포트
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
```

```python
## 특성과 정답 선언
x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])

## 텐서 플로우 모델 객체 선언
model = Sequential()
## 덴스 레이어 추가(레이어 1개 , input 특성 2개, 분석 방법 : 선형 회귀)
model.add(Dense(1, input_dim = 2, activation = 'linear'))
## 오차 수정을 위해 경사하강법(sgd), 오차 계산으로 평균 제곱 오차(mse) 사용
model.compile(optimizer = 'sgd', loss = 'mse')
## 오차 최소화 과정 2000번 반복 훈련
model.fit(x,y,epochs = 2000)
```

```python
## 공부시간이 7시간, 과외 시간이 4시간일 때 점수 예측후 확인
hour = 7
private_class = 4
prediction = model.predict([[hour, private_class]])
print('%.f시간을 공부하고, %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점 입니다.' % (hour, private_class, prediction))
```
