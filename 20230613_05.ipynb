```python
import pandas as pd
import seaborn as sns
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.info()
```

```python
titanic.head()
```

```python
titanic['embarked'].value_counts()
```

```python
titanic['embarked'].fillna('S', inplace=True)

titanic.info()
```

```python
titanic['deck'].value_counts()
```

```python
df_titanic = titanic.iloc[:, [0,2,3,6,7,8,14]]
```

```python
df_titanic.head()
```

```python
df_titanic.info()
```

```python
df_titanic['age'].mean()
```

```python
df_titanic['age'].fillna(df_titanic['age'].mean(), inplace=True)
```

```python
df_titanic.info()
```

```python
df_titanic
```

```python
df_sex = pd.get_dummies(df_titanic['sex'])

df_sex
```

```python
df_embarked = pd.get_dummies(df_titanic['embarked'])

df_embarked
```

```python
df_class = pd.get_dummies(df_titanic['class'])

df_class
```

```python
df_alone = pd.get_dummies(df_titanic['alone'])

df_alone
```

```python
df_titanic.drop(['sex', 'embarked', 'class', 'alone'], axis= 1, inplace=True)
```

```python
df_titanic.info()
```

```python
df_titanic
```

```python
df_titanic = pd.concat([df_titanic, df_sex, df_embarked, df_class, df_alone],axis= 1)
```

```python
df_titanic.info()
```

```python
df_titanic.shape
```

```python
df_titanic.head()
```

```python
X_titanic = df_titanic.iloc[:,1:]
y_titanic = df_titanic.iloc[:,0]

print(X_titanic.shape, y_titanic.shape)
```

```python
X_titanic.astype(float)
X_titanic.loc['female': ] = X_titanic.loc['female': ].astype(float)
```

```python
X_titanic.info()
```

```python
## 모델 객체 선언
model = Sequential()
## 3개의 layer생성
## 1. 노드 : 24개, input데이터 : 60개, 활성화 함수 : 렐루함수
## 2. 노드 : 10개, 활성화 함수 : 렐루함수
## 3. 노드 : 1개, 활성화 함수 : 시그모이드 함수
model.add(Dense(24, input_dim = 12, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
## 모델 컴파일
## 손실 함수 : binary_crossentropy, 옵티마이저 : adam
model.compile(loss='binary_crossentropy', optimizer= 'adam', metrics=['accuracy'])
```

```python
## 모델 훈련
history = model.fit(X_titanic,y_titanic,epochs=2000)
```
